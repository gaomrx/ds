#  Age of Abalones

<body>
Abalone is a common name for a group of small and generally edible marine snails. Estimating an abalone's age can be attained by sawing through its shell, staining the shell and counting the number of rings present. However, this can be a boring and time consuming task. Hence, figuring out a better approach by taking into account other physical features of the abalone can be more efficient.

The age of abalones can be estimated as 1.5 + number of rings present. It is convenient to group abalones into three age categories based on number of rings: young (1-8 rings), medium (9-10) and old (11+). 

</body>

Looking at the distribution of rings, it seems odd that a 10 year old abalone can still be classified as an infant.
After reading a little into abalone sexing, "I" should really stand for "indetermined" rather than infant. 
Furthermore, there is actually [no clear signs of sexual dimorphism](http://www.sciencedirect.com/science/article/pii/S0165783608002440) between male and female as seen in the distribution of physical traits.
Without any distinction for "gender", all the abalones will be grouped together for the remaining analysis.

```python
# sanity check for whole weight versus sum of shell, shucked and gut weight
# the whole weight should be greater or equal to the sum of the other weights

print sum(1 for idx in df.index if df.ix[idx]["whole_weight"] < (df.ix[idx]["shucked_weight"] + df.ix[idx]["viscera_weight"] + df.ix[idx]["shell_weight"]))
```

<body> Another oddity, 155 abalones apparently weigh less than the sum of their other parts. Let's only consider whole weight from now on as the people gathering the data would have more likely misrecorded the to other weights rather than the whole weight. 
<br> <br/>
For the acutal task of predicting age categories, a regression problem wrapped as a classification problem, we need to note that the physical traits provided here are highly correlated (eg: heavier abalones also tend to be larger). Instead of using the usual multiple linear regression which is susceptible to multicolinnearity, let's try something more robust like support vector regression.
</body>

```python
abdf = df[["length", "diameter", "height", "whole_weight"]]

# age category, 0 = young, 1 = medium, 2 = old
age = []
for idx in df.index:
    if df.ix[idx]["rings"] < 9:
        age.append(0)
    elif 9 < df.ix[idx]["rings"] < 11:
        age.append(1)
    else:
        age.append(2)
print len(age)
```

```python
# formatting dataframes as input to SVR
X = abdf.as_matrix()
y = df[["rings"]].as_matrix().ravel()
print y
```

```python
# classify tnto young(0), medium(1) or old(2) category based on number of rings
def classify_age(predict_rings):
    predict_age = []
    for i in predict_rings:
        if i < 9:
            predict_age.append(0)
        elif 9 < i < 11:
            predict_age.append(1)
        else:
            predict_age.append(2)
    return predict_age
```

```python
temp_y = classify_age(y)
print sum(np.array(temp_y)-np.array(age))
```

```python
from sklearn.svm import SVR
```

```python
# trying out different SVR with mostly default settings
svr_rbf = SVR(kernel='rbf', C=1e3)
svr_lin = SVR(kernel='linear', C=1e3)
y_rbf = svr_rbf.fit(X, y).predict(X)
y_lin = svr_lin.fit(X, y).predict(X)
```

```python
# predicted ages
print y_rbf, y_lin
```

```python
# coefficient of determination with mostly default settings
print svr_rbf.score(X,y), svr_lin.score(X,y)
```

```python
rbf_age = classify_age(y_rbf)
```

```python
lin_age = classify_age(y_lin)
```

```python
# confusion matrix
rbf_cm = sklearn.metrics.confusion_matrix(age, rbf_age)

print rbf_cm

#       predicted young, medium, old
# true
# young
# medium
# old
```

```python
# SVR with RBF 
print sklearn.metrics.classification_report(age, rbf_age, target_names=["young", "medium", "old"])
```

```python
# confusion matrix
lin_cm = sklearn.metrics.confusion_matrix(age, lin_age)

print lin_cm

#       predicted young, medium, old
# true
# young
# medium
# old
```

```python
# linear SVR
print sklearn.metrics.classification_report(age, lin_age, target_names=["young", "medium", "old"])
```

<body>

In summary, the coefficient of determination for the linear and RBF SVR were 0.30 and 0.35 respectively, which are kind of low, but not suprising as we only considerd four physical features for predicting age. This also reflects in the  accuracy metrics for both models, although one may tweak the settings to get better estimators. But then, these are just abalones, so does it really matter what their ages are as long as they taste good? Food for thought I guess.

</body>

<body> 

Seeking a more parsimonious prediction for number of rings with insight from knowing that shell diameter is directly related to the number of rings.

</body>


```python
# pairwise correlation between rings and other physical features
abdf2 = df[["length", "diameter", "height", "whole_weight", "rings"]]
cor = abdf2.corr()
cor 
```

```python
# covariance matrix for physical features and condition number test
cov = cor * cor
cov
```

```python
# condition number of covariance matrix is above 30, very certain that multicolinearity will be a problem if doing MLS
# alternative is to try ridge regression
np.linalg.cond(cov)
```

```python
# ridge regression
# alpha controls the amount of "shrinkage" in coefficients of ridge regression
# alpha -> 0, ridge regression coefficients ~ MLS coeffficients; alpha -> infinity, ridge regression coefficients -> 0

from sklearn.linear_model import Ridge

# ridge regression helper function to show results
def rr_results(alpha):
    X = abdf.as_matrix()
    y = df[["rings"]].as_matrix().ravel()
    true_age = classify_age(y)
    rr = Ridge(alpha = alpha)
    rr.fit(X,y)
    rr_y = rr.predict(X)
    predict_age = classify_age(rr_y)
    print "ridge regression with alpha : ", alpha
    print "coefficient of determination : " , rr.score(X,y)
    print "confusion matrix : \n", sklearn.metrics.confusion_matrix(true_age, predict_age)
    #       predicted young, medium, old
    # true
    # young
    # medium
    # old
    
    print "\n classification metrics : \n ", sklearn.metrics.classification_report(true_age, predict_age, target_names=["young", "medium", "old"])
    return
```

```python
rr_results(0) # slightly better than SVR
```

```python
rr_results(1) # slightly better than SVR
```

```python
rr_results(0.1) # slightly better than SVR
```

```python
# linear regression on rings with diamemter

def linreg_results(var):
    print "linear regression explanatory variable : ", var
    from sklearn.linear_model import LinearRegression as linreg
    X = abdf[[var]].as_matrix()
    y = df[["rings"]].as_matrix().ravel()
    true_age = classify_age(y)
    lr = linreg()
    lr.fit(X,y)
    predict_age = classify_age(lr.predict(X))
    print "coefficient of determination : ", lr.score(X,y)
    print "confusion matrix : \n", sklearn.metrics.confusion_matrix(true_age, predict_age)
    #       predicted young, medium, old
    # true
    # young
    # medium
    # old
    
    print "\n classification metrics : \n ", sklearn.metrics.classification_report(true_age, predict_age)
    return

                                                                                   
```

```python
linreg_results("height")
```

```python
linreg_results("diameter")
```

```python
linreg_results("whole_weight")
```

```python
linreg_results("length") 
```

<body>
It turns out that ridge regression performed better than the SVR and simple OLS regression approaches, but none of them were really great at predicticting age judging by the F1 scores and coefficient of determination. The insight that shell diameter is directly related to numbers of rings is still true and more stongly correlated than other physical features like weight or length.
